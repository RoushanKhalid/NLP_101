```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('spam.csv', encoding='latin-1')
data
```

```python
data = data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1)
data
```

```python
!pip install nltk
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer
count_vectorizer = CountVectorizer()

# Fit and transform the 'message' column
count_matrix = count_vectorizer.fit_transform(data['message'])

# Now apply TfidfTransformer to the count matrix
X = tfidf_transformer.fit_transform(count_matrix).toarray()
X
```

```python
# Access the vocabulary from the CountVectorizer
vocabulary = count_vectorizer.get_feature_names_out()
vocabulary
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X, data['label'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, data['label'], test_size=0.3, random_state=0)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
print("Accuracy:", score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
